```python
import seaborn as sns
```

Key Features of Seaborn

**Built on top of Matplotlib**: Seaborn uses Matplotlib as its backend, so you can use Matplotlib's functionality and customization options.<br/>
**High-level interface**: Seaborn provides a simple and intuitive API for creating complex plots, making it easier to focus on the data rather than the plotting details.<br/>
**Statistical graphics**: Seaborn is designed for statistical data visualization, with a focus on showing relationships, distributions, and correlations.<br/>
**Integration with Pandas**: Seaborn is designed to work seamlessly with Pandas DataFrames, making it easy to visualize and explore datasets.
Common Seaborn Plots

**relplot**: Relational plots (like the one in your code) show the relationship between two variables.<br/>
**catplo**t: Categorical plots show the relationship between a categorical variable and a numerical variable.<br/>
**distplot**: Distribution plots show the distribution of a single variable.<br/>
**pairplot**: Pair plots show the relationships between multiple variables.<br/>
**heatmap**: Heatmaps show the relationship between two variables as a 2D matrix.<br/>
Tips for Using Seaborn

Start with a simple plot: Begin with a basic plot and then customize it to your needs.
Use the data parameter: Pass your dataset as a Pandas DataFrame to the data parameter to take advantage of Seaborn's automatic data handling.
Customize with options: Use Seaborn's various options (e.g., hue, style, size) to add additional dimensions to your plots.
Explore different plot types: Try out different plot types to find the one that best suits your data and question.

```python
sns.set_theme()
tips=sns.load_dataset("tips")
tips
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    style="sex",
    size="size"

)
```

```python
sns.catplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex",
    kind="bar"

)
```

```python
sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    col="time"

)
```

```python
sns.jointplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker"



)
```

```python
# tips=tips.drop(columns=["day","time"])
tips=tips.drop(columns=["sex","smoker"])
tips.corr()
```

```python
sns.heatmap(
    data=tips.corr(),
    annot=True,
    cmap="coolwarm"
)
```
